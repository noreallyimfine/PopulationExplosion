```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population = pd.read_csv("population_total.csv")

print(population.shape)

population.head()
```

```python
population = population.loc[:, :'2020']
population.head()
```

```python
fertility = pd.read_csv("children_per_woman_total_fertility.csv")

print(fertility.shape)

fertility.head()
```

```python
avg_fert = fertility.describe().loc['mean', :'2020']
```

```python
X = avg_fert.index.tolist()
x_labels = [x for x in X if int(x) % 25 == 0 or x == '2020']

plt.plot(
    X,
    avg_fert.values
    )

plt.xticks(x_labels, x_labels);
plt.title("Fertility rate stays above 5 until after 1975");
plt.ylabel("Children per woman");
plt.xlabel("Year");
```

```python
sum_pop = population.sum()[1:]
```

```python
X = sum_pop.index.tolist()
x_labels = [x for x in X if int(x) % 25 == 0 or x == '2020']

plt.plot(
    X,
    sum_pop.values
    )

plt.xticks(x_labels, x_labels);
plt.ylabel("World population (in billions)");
plt.xlabel("Year");
plt.title("World population crosses 5 billion around 1975");
```
